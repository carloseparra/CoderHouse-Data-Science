## Obtener listado de ordenes de compra generadas en un día especifico

```python
import requests
import json
import pandas as pd
url='https://api.mercadopublico.cl/servicios/v1/publico/ordenesdecompra.json?fecha=08032023&ticket='+ api_key
df= requests.get(url)
print(df)
df.text
```

```python
jsondata=json.loads(df.text)
df2=pd.DataFrame.from_dict(jsondata)
print('Número de filas:', len(df2))
display (df2.head())
print("-*-*-*-*-*-*-*-*-*-* D E S E M P A Q U E T A R   J S O N -*-*-*-*-*-*-*-*-*-*-*-*")
# Desempaquetar Listado de OC
df3 = pd.json_normalize(jsondata, record_path=['Listado'], meta = ['Cantidad', 'Version'])
df3
```

```python
# OTRA MANERA DE CREAR EL URL CON FORMAT
url ='https://api.mercadopublico.cl/servicios/v1/publico/ordenesdecompra.json?fecha=08032023&ticket={0}'\
  .format(api_key)
# request data
response = requests.request("GET", url)
# view status code
print(response.status_code)
#response.text
#response.json()
```

## Obtener detalle de una orden de compra especifica
